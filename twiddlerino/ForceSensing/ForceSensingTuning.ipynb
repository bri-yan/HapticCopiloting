```python
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
data = open("TorqueDataNoResistance.txt")

expectedTorque = []
totalTorque = []

for line in data:
    nums = re.findall(r"[-+]?\d*\.\d+|\d+", line)
    expectedTorque.append(float(nums[0]))
    totalTorque.append(float(nums[1]))

expectedTorque = np.array(expectedTorque)
totalTorque = np.array(totalTorque)

plt.plot(expectedTorque)
plt.plot(totalTorque)
plt.title("Expected Torque + Total Torque, No User Resistance")
plt.xlabel("Time (x35ms)")
plt.ylabel("Torque (Ncm)")
plt.legend(["Expected", "Total"])
```

```python
data = open("TorqueDataSomeResistance.txt")

expectedTorque = []
totalTorque = []

for line in data:
    nums = re.findall(r"[-+]?\d*\.\d+|\d+", line)
    expectedTorque.append(float(nums[0]))
    totalTorque.append(float(nums[1]))

expectedTorque = np.array(expectedTorque)
totalTorque = np.array(totalTorque)

plt.plot(expectedTorque)
plt.plot(totalTorque)
plt.title("Expected Torque + Total Torque, Some User Resistance")
plt.xlabel("Time (x35ms)")
plt.ylabel("Torque (Ncm)")
plt.legend(["Expected", "Total"])
```

```python
userTorque = totalTorque - expectedTorque

plt.plot(userTorque)
plt.title("External User Torque, Some User Resistance")
plt.xlabel("Time (x35ms)")
plt.ylabel("Torque (Ncm)")
```
