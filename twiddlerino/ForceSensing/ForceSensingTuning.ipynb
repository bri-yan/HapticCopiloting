```python
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
data = open("data1.txt")

expectedCurrent = []
actualCurrent = []

for line in data:
    nums = re.findall(r"[-+]?\d*\.\d+|\d+", line)
    expectedCurrent.append(float(nums[0]))
    actualCurrent.append(float(nums[1]))

expectedCurrent = np.array(expectedCurrent)
actualCurrent = np.array(actualCurrent)

plt.plot(expectedCurrent)
plt.plot(actualCurrent)
plt.title("Expected Current vs. Actual Current")
plt.xlabel("Time")
plt.ylabel("Current")
plt.legend(["Expected", "Actual"])
```
